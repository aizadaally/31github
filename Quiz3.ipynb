```python
import random


graph = {
  '5': ['3', '7'],
  '3': ['2', '4'],
  '7': ['8'],
  '2': [],
  '4': ['8'],
  '8': []
}

visited = set()

def dfs(visited, graph, node, target):
    if node not in visited:
        print(node)
        visited.add(node)
        if node == target:
            return True
        for neighbour in graph[node]:
            if dfs(visited, graph, neighbour, target):
                return True
    return False

def bfs(visited, graph, node, target):
    visited.append(node)
    queue = [node]

    while queue:
        m = queue.pop(0)
        print(m, end=" ")
        if m == target:
            return True
        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)


random_target = random.choice(list(graph.keys()))
print(f"Randomly selected target node: {random_target}")

print("Following is the Breadth-First Search")
search_result_bfs = bfs([], graph, '5', random_target)
if search_result_bfs:
    print("\nSearch is Successful")
else:
    print("\nSearch is Unsuccessful")

print("\nFollowing is the Depth-First Search")
visited = set()
search_result_dfs = dfs(visited, graph, '5', random_target)
if search_result_dfs:
    print("Search is Successful")
else:
    print("Search is Unsuccessful")




```
